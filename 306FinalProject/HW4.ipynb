## Load Data to Mysql Tables

```python



import csv
from mysql.connector import IntegrityError

#Language ıcın 
def populate_table1(db_connection, db_cursor, insert_query, csv_file):
    with open(csv_file, 'r', newline='') as file:
         

        for line in file:
            row = line.strip().split(';')

            imdb_id = row[0].strip() if row[0] else None
            language = row[1].strip() if row[1] else None
            country = row[2].strip() if row[2] else None

            #print(f"Inserting row: {imdb_id}, {language}, {country}")

            try:
                db_cursor.execute(insert_query, (imdb_id, language, country))
            except mysql.connector.IntegrityError:
                update_query = "UPDATE LANGUAGE_FINAL15 SET language = %s, country = %s WHERE imdb_id = %s"
                db_cursor.execute(update_query, (imdb_id, language, country))

    db_connection.commit()

#Movıe ıcın
def populate_table2(db_connection, db_cursor, insert_query, csv_file):
    with open(csv_file, 'r', newline='') as file:
        csv_data = csv.reader(file, delimiter=';')
        next(csv_data)  # Skip the header row

        for row in csv_data:
            row = [item.strip() if item else None for item in row]  # Remove leading/trailing whitespaces

            imdb_id, title, year, rating, runtime, genre, released, awards, lastupdated, type = row

            #print(f"Inserting row: {imdb_id}, {title}, {year}, {rating}, {runtime}, {genre}, {released}, {awards}, {lastupdated}, {type}")

            try:
                db_cursor.execute(insert_query, (imdb_id, title, year, rating, runtime, genre, released, awards, lastupdated, type))
            except mysql.connector.IntegrityError:
                continue

    db_connection.commit()

#Crıtıc movıe detaıl
def populate_table3(db_connection, db_cursor, insert_query, csv_file):
    with open(csv_file, 'r', newline='') as file:
         
        for line in file:
            row = line.strip().split(';')

            critic_id = row[0].strip() if row[0] else None
            imdb_rating = row[1].strip() if row[1] else None
            imdbVotes = row[2].strip() if row[2] else None

            #print(f"Inserting row: {critic_id}, {imdb_rating}, {imdbVotes}")

            try:
                db_cursor.execute(insert_query, (critic_id, imdb_rating, imdbVotes))
            except mysql.connector.IntegrityError:
                update_query = "UPDATE CRITICS_DETAIL SET critic_id = %s, imdb_rating = %s ,imdbVotes = %s"
                db_cursor.execute(update_query, (critic_id, imdb_rating, imdbVotes))
    db_connection.commit()


#crıtıc relatıon ıcıın
def populate_table4(db_connection, db_cursor, insert_query, csv_file):
    with open(csv_file, 'r', newline='') as file:
         
        for line in file:
            row = line.strip().split(';')

            imdbID = row[0].strip() if row[0] else None
            critic_id = row[1].strip() if row[1] else None
    

            #print(f"Inserting row: {imdbID}, {critic_id}")

            try:
                db_cursor.execute(insert_query, (imdbID, critic_id))
            except mysql.connector.IntegrityError:
                update_query = "UPDATE CRITICS_MAIN SET imdbID = %s, critic_id = %s "
                db_cursor.execute(update_query, (imdbID, critic_id))
    db_connection.commit()




#story relatıon ıcın 
def populate_table5(db_connection, db_cursor, insert_query, csv_file):
    with open(csv_file, 'r', newline='') as file:
         
        for line in file:
            row = line.strip().split(';')

            imdbID = row[0].strip() if row[0] else None
            storyID = row[1].strip() if row[1] else None
    

            #print(f"Inserting row: {imdbID}, {storyID}")

            try:
                db_cursor.execute(insert_query, (imdbID, storyID))
            except mysql.connector.IntegrityError:
                update_query = "UPDATE STORY_MAIN SET imdbID = %s, storyID = %s "
                db_cursor.execute(update_query, (imdbID, storyID))
    db_connection.commit()



#story detaıl ıcın 
def populate_table6(db_connection, db_cursor, insert_query, csv_file):
    with open(csv_file, 'r', newline='') as file:
         
        for line in file:
            row = line.strip().split(';')

            storyID = row[0].strip() if row[0] else None
            poster = row[1].strip() if row[1] else None
            plot = row[2].strip() if row[2] else None
            fullplot = row[2].strip() if row[2] else None

            #print(f"Inserting row: {storyID}, {poster} ,{plot}, {fullplot}  ")

            try:
                db_cursor.execute(insert_query, (storyID, poster ,plot,fullplot))
            except mysql.connector.IntegrityError:
                update_query = "UPDATE STORY_DETAIL SET storyID = %s, poster = %s ,plot = %s,fullplot = %s "
                db_cursor.execute(update_query, (storyID, poster ,plot,fullplot))
    db_connection.commit()



#CAST DETAIL
def populate_table7(db_connection, db_cursor, insert_query, csv_file):
    with open(csv_file, 'r', newline='', encoding='utf-8') as file:

         
        for line in file:
            row = line.strip().split(';')

            castID = row[0].strip() if row[0] else None
            cast = row[1].strip() if row[1] else None
            writer = row[2].strip() if row[2] else None
            director = row[2].strip() if row[2] else None

            #print(f"Inserting row: {castID}, {cast} ,{writer}, {director}  ")

            try:
                db_cursor.execute(insert_query, (castID, cast ,writer,director))
            except mysql.connector.IntegrityError:
                update_query = "UPDATE CAST_DETAIL SET storyID = %s, cast = %s ,writer = %s,director = %s "
                db_cursor.execute(update_query, (castID, cast ,writer,director))
    db_connection.commit()


#MOVİE CAST
def populate_table8(db_connection, db_cursor, insert_query, csv_file):
    with open(csv_file, 'r', newline='') as file:
         
        for line in file:
            row = line.strip().split(';')

            imdbID = row[0].strip() if row[0] else None
            castID = row[1].strip() if row[1] else None
    

            #print(f"Inserting row: {imdbID}, {castID}")

            try:
                db_cursor.execute(insert_query, (imdbID, castID))
            except mysql.connector.IntegrityError:
                update_query = "UPDATE CAST_RELATION SET imdbID = %s, castID = %s "
                db_cursor.execute(update_query, (imdbID, castID))
    db_connection.commit()

```

```python
db_cursor.execute("""
    CREATE TABLE LANGUAGE_FINAL15 (
        imdb_ID VARCHAR(45) PRIMARY KEY,
        language VARCHAR(255),
        country VARCHAR(255)
    )
""")

insert_movie = (
    "INSERT INTO LANGUAGE_FINAL15(imdb_id, language, country) "
    "VALUES (%s, %s, %s)"
)

populate_table1(db_connection, db_cursor, insert_movie, "LanguageMovie.csv")

                                                        
```
